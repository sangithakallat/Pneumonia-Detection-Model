```python
from tensorflow.keras.models import load_model

# Load the saved model
model = load_model('C:/Users/Admin/Documents/SEM3/Project/Models/model4_fine_tuned_45.keras')
```

```python
import os
import numpy as np
from tensorflow.keras.preprocessing import image
img_width = 224
img_height = 224
```

```python
import os
import csv
import numpy as np
from tensorflow.keras.preprocessing import image

# Load the trained model
#model = ...  # Load your trained model here

# Define the path to the folder containing the images
image_folder = r'C:\Users\Admin\Documents\SEM3\Project\Chest_Xray\test\PNEUMONIA'

# Define the path to the CSV file
csv_file = 'C:/Users/Admin/Documents/SEM3/Project/Prediction_Results/pneumonia_prediction_six.csv'

# Create or open the CSV file in write mode
with open(csv_file, 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Image', 'Prediction'])

    # Get a list of image files in the folder
    image_files = os.listdir(image_folder)

    # Iterate over the image files and make predictions
    for filename in image_files:
        # Load and preprocess the image
        img_path = os.path.join(image_folder, filename)
        img = image.load_img(img_path, target_size=(img_width, img_height))
        img = image.img_to_array(img)
        img = np.expand_dims(img, axis=0)
        img = img / 255.0  # Normalize the image

        # Make the prediction
        prediction = model.predict(img)
        predicted_class = 'Pneumonia' if prediction > 0.5 else 'Normal'

        # Write the prediction result to the CSV file
        writer.writerow([filename, predicted_class])
```
